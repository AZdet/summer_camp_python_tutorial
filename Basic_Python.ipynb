## Images
In this course, we are dealing with images most of the time. In the computer, images are represented as a 3D array. It typically has the shape (H, W, 3). The first two parameters H, W are the height and width of the image. These two parameters determine how many pixels in the image spatially. 3 corresponds to R(red), G(green), B(blue) channel of the image. Every color we perceive can be composed as a combination of red, green and blue color. We use 3 numbers ranging from 0 to 255 to represent saturation of each color. Below are a few RGB values for common colors

![img](color_table.png)

You can play around with more RGB values [here](https://www.w3schools.com/colors/colors_rgb.asp).

````python
import cv2
import matplotlib.pyplot as plt

def show_image(brg_img):
    rgb_img = cv2.cvtColor(brg_img, cv2.COLOR_BGR2RGB)

    plt.imshow(rgb_img)
    plt.show()

def show_image_channel(brg_img, channel):
    img_channel = cv2.cvtColor(brg_img, cv2.COLOR_BGR2RGB)
    if channel == 0:
        img_channel[:, :, 1] = 0
        img_channel[:, :, 2] = 0
    if channel == 1:
        img_channel[:, :, 0] = 0
        img_channel[:, :, 2] = 0
    if channel == 2:
        img_channel[:, :, 0] = 0
        img_channel[:, :, 1] = 0
    plt.imshow(img_channel)
    plt.show()

def show_image_content(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    fig, ax = plt.subplots(1, 3, figsize=(20,20))
    limit = 10
    channel = ['R', 'G', 'B']
    for ch in range(3):
        ax[ch].imshow(img[-limit:, -limit:, ch], cmap='gray')
        for i in range(limit):
            for j in range(limit):
                ax[ch].text(j, i, img[-(limit-i), -(limit-j), ch], ha='center', va='center', color='r')
        ax[ch].set_title(channel[ch] + ' channel')
    fig.tight_layout()
    plt.show()
````

````python
# How to load images and show them
# cv2.imread needs to be given the image name
cat = cv2.imread('cat.jpg')
show_image(cat)
````

````python
#  We can view the dimensions of the image using .shape
#  The dimensions are height(rows), width(columns), number of channels
cat.shape
````

If we examine part of the image(here we take the last 100 pixels), we can see that there is a value between 0 and 255 at each location and there is 3 channels. The values in 3 channels can be different.

````python
show_image_content(cat)
````

## Image Class
To summarize everything, we will create a python class for images that is able to do everything we have seen before but packages it all into one thing that allows for consecutive manipulations.

````python
class Image():
    
    def __init__(self, image_name=None, image_data=None):
        """
        Takes in either a filename to be loaded and stored, or 
        if the image data is already provided, just store the data
        You only need to provide one of:
            image_name
            image_data
        Don't need both
        """
        # load the image name and assign the image array to self.image_array
        # Check if the image_name or image_data was provided by using if, else
        
        if not image_name == None:
            self.image_array = cv2.imread(image_name)
        else:
            self.image_array = image_data
        

    
    def fill(self, starting_coordinates, end_coordinates, value):
        """
        Fills a rectangular part of the image with a specific color
        returns a new Image object with the modified image
        The coordinates expected are of the form (x, y) and not (y, x) or column, row
        the .copy() function is used so that the original image will not be modified
        ex:
            Image.fill((5,5), (10, 10), 0)
            will fill the image starting at coordinates 5,5 and ending at 10,10
            with a black square
        """
        # create variables called: start_x, start_y, end_x, end_y
        # Then use the slice operation to fill with the given value.
        new_image_array = self.image_array.copy()
        start_x = starting_coordinates[0]
        start_y = starting_coordinates[1]
        end_x = end_coordinates[0]
        end_y = end_coordinates[1]
        new_image_array[start_y:end_y, start_x:end_x] = value
        new_image_obj = Image(image_data=new_image_array)
        return new_image_obj

    def crop(self, starting_coordinates, end_coordinates):
        """
        Crops a rectangular part of the image
        returns a new Image object with the modified image
        The coordinates expected are of the form (x, y) and not (y, x) or column, row
        ex:
            Image.crop((5,5), (10, 10))
            will fill image_array with only the rectangle starting at 5,5 and ending at 10,10
        """
        # create variables called: start_x, start_y, end_x, end_y
        # Don't need to create a copy this time since we aren't filling with new values
        # Use the slice operation to crop
        
        start_x = starting_coordinates[0]
        start_y = starting_coordinates[1]
        end_x = end_coordinates[0]
        end_y = end_coordinates[1]
        new_image_array = self.image_array[start_y:end_y, start_x:end_x]
        new_image_obj = Image(image_data=new_image_array)
        return new_image_obj
   
    def paste(self, image_obj, starting_coordinate):
        """
        Pastes given image into the current image
        returns a new Image object with the modified image

        image_obj is an Image object, so it is an object you already created
        The coordinates expected are of the form (x, y) and not (y, x) or column, row
        """
        # create variables called: start_x, start_y
        # You can find the end of the slice by looking at the dimensions of the image to be painted in
        new_image_array = self.image_array.copy()
        start_x = starting_coordinate[0]
        start_y = starting_coordinate[1]
        image_height = image_obj.image_array.shape[0]
        image_width = image_obj.image_array.shape[1]
        new_image_array[start_y:start_y+image_height, start_x:start_x+image_width] = image_obj.image_array
        new_image_obj = Image(image_data=new_image_array)
        return new_image_obj
    
   
    def save(self, save_name):
        cv2.imwrite(save_name, self.image_array)
    
    def display(self):
        rgb_img = cv2.cvtColor(self.image_array, cv2.COLOR_BGR2RGB)
        plt.imshow(rgb_img)
        plt.show()
        
````

````python
thanos = Image(image_name="thanos.jpg")
thanos.display()
````

````python
tb12 = Image(image_data=cv2.imread("tom_brady.jpg"))
tb12.display()
````

````python
six_SB_rings = thanos.paste(tb12, (300, 30))
six_SB_rings.display()
````
